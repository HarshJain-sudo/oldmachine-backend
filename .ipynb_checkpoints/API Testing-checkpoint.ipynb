```python
import requests
import json
```

```python
base_url = "http://127.0.0.1:8000/api/marketplace"
access_token = ""
```

```python
def make_api_call(url: str, payload, access_token=None):
        try:
            HEADERS = {"Content-Type": "application/json"}
            response = requests.post(url, json=payload, headers=HEADERS, timeout=10)
            
            print(f"Status Code: {response.status_code}")
            
            if response.status_code == 200:
                data = response.json()
                print(data)
                
                return None
            else:
                print(f"Login/Sign Up failed!")
                print(response.json())
                return None
            
        except Exception as e:
            print(f"Error: {e}")
            return None

    
```

### SignIn SignUp API

```python
payload = {
    "phone_number": "7467885047",
    "country_code": "+91"
}

url = f"{base_url}/login_or_sign_up/v1/"
make_api_call(url, payload)
```

### Verify OTP

```python
payload = {
    "phone_number": "7467885047",
    "otp": "000000"
}

url = f"{base_url}/verify_otp/v1/"
make_api_call(url, payload)
```

### Get Categories API

```python
payload = {
    "phone_number": "7467885047",
    "otp": "000000"
}
limit = 10
offset = 0

url = f"{base_url}/categories_details/get/v1/?limit={limit}&offset={offset}"
make_api_call(url, payload)
```
